```python
documents = [
{"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
{"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
{"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
]
directories = {
"1": ["2207 876234", "11-2"],
"2": ["10006"],
"3": []
}
```

Пользователь по команде “p” может узнать владельца документа по его номеру
Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится
Пользователь по команде “l” может увидеть полную информацию по всем документам
Пользователь по команде “as” может добавить новую полку
Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)
Пользователь по команде “ad” может добавить новый документ в данные
Пользователь по команде “d” может удалить документ из данных
Пользователь по команде “m” может переместить документ с полки на полку


```python
def docnum(ins):               #возвращает владельца по номеру документа
     for doc in documents:
        if doc["number"] ==ins:
            return doc["name"]
        
def dirnum(a):                 #возвращает номер полки по номеру документа
    for i in directories.items():
        for k in i[1]:
            if k == a:
                return (i[0])
def type_(t):                  #возвращает тип по номеру документа 
    for doc in documents:
        if doc["number"] ==t:
            return doc["type"]
        
def all_():                   #печатает все, что есть
    for doc in documents:
                  print(f"№: {doc['number']}, тип: {type_(doc['number'])}, владелец:{docnum(doc['number'])}, полка: {dirnum(doc['number'])}" )
        
        
        
def num_of_dirs():             #возвращает количество доступных полок
    k = []
    for i in directories.items():
                k.append(i[0])
    return k

def del_shelf(sh_num):         #удаляет полку по номеру(или не удаляет, если такой нет или она занята
    import copy
    copy_directories = copy.copy(directories)
    if sh_num not in num_of_dirs(): 
        return 2
    else:
        for i in copy_directories.items():   
                 if i[0] == sh_num:
                        if i[1] == []:
                                del(directories[i[0]])
                                return 1       
                        else:
                                return 0
def new_line(a,b,c,d):                                #добавляет документ и полку
    if d in num_of_dirs():                      
             documents.append({"type":a,"number":b,"name":c})
             for i in directories.items():
                if i[0] == d:
                    i[1].append(b) 
                    return 1
    else:
        return 0
    
def new_line_dir(a,b):                                    #добавляет докумет в полку
    for i in directories.items():
                if i[0] == a: #номер полки
                    i[1].append(b) #номер документа
                   
    
    
    
def del_line(ins):           #удаляет документ по номеру отовсюду
    if docnum(ins) != None:
        import copy
        сopy_directories = copy.copy(directories)
        for i in directories.items(): 
                         if ins in i[1]:
                              for k in i[1]:
                                     if ins == k:
                                         del(i[1][i[1].index(k)])
 
        copy_documents = copy.copy(documents)
        for doc in copy_documents:
               if doc["number"] ==ins:
                    del(documents[copy_documents.index(doc)])
                    return 1
    else:
        return 0
    
def del_line_dir(shelf,num):               #перемещает документ с полки на полку
    if shelf not in num_of_dirs():
        print("Такой полки не существует. Текущий перечень полок: ")
    else:
        if docnum(num) == None:
            print("Документ не найден в базе.Текущий список документов: ")
        else:
            сopy_directories = copy.copy(directories)
            for i in directories.items(): 
                           if num in i[1]:
                                for k in i[1]:
                                      if num == k:
                                            del(i[1][i[1].index(k)])
                                            new_line_dir(shelf,num)
                                            print("Документ перемещен.Текущий список документов: ")
     
    
    
resp = ""
while resp  != "q":
    resp = input("введите команду: ")
    print("")
    if resp == "q":
        print("game over")
        break
        
        
    elif resp == "p":
        n=input("Введите номер документа:")
        if docnum(n) != None:
            print(docnum(n))
        else:
            print("Документ не найден в базе")
    elif resp == "s":
        a=input("Введите номер документа:")   
        if dirnum(a) != None:  
            print(f"Документ хранится на полке:{dirnum(a)} ")
        else:
            print("Документ не найден в базе")
    elif resp == "l": 
         all_()
    elif resp == "as":
        shelf=input("Введите номер полки:") 
        if shelf in num_of_dirs():
            print(f"Такая полка уже существует. Текущий перечень полок:{list(num_of_dirs())} ")
        else:
            directories.update({shelf:[]})
            print(f"Полка добавлена. Текущий перечень полок:{list(num_of_dirs())}")
    elif resp == "ds":
        shelf_to_del=input("Введите номер полки:")
        k = del_shelf(shelf_to_del)
        if k == 0:
            print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {list(num_of_dirs())}")
        elif k == 2:
            print(f"Такой полки не существует. Текущий перечень полок: {list(num_of_dirs())}")
        elif k == 1:
            print(f"Полка удалена. Текущий перечень полок: {list(num_of_dirs())}")
    elif resp == "ad":
        num = input("Введите номер документа: ")
        typ = input("Введите тип документа: ")
        owner = input("Вдите владельца документа: ")
        shelf_ = input("Введите полку для хранения: ")
        k=new_line(typ,num,owner,shelf_)
        if k == 1:
            all_()
        else:
            print("Tакой полки не существует. Добавьте полку командой as. Текущий список документов:")
            all_()
    elif resp == "d":
        num = input("Введите номер документа: ")
        k = del_line(num)
        if k == 1:
            print("Документ удален.Текущий список документов:")
            all_()
        else: 
            print("Документ не найден в базе.Текущий список документов:")
            all_()
    elif resp == "m":  
        num = input("Введите номер документа: ")
        shelf = input("Введите номер полки: ")
        del_line_dir(shelf,num)
        all_()
    else:#
        print("команда не существует")

```
